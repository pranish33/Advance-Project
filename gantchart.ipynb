```python
# Let's reverse the Gantt chart tasks order to match the original request.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for the Gantt chart
tasks = [
    "Proposal Submission",
    "Requirements Gathering",
    "Research on Algorithms",
    "UI/UX Design",
    "Project Flow Design",
    "Database Design",
    "System Architecture Design",
    "Initial Design Submission",
    "Initial Testing",
    "Coding (1)",
    "Bug Fixing (If needed)",
    "Final Testing",
    "Professionalism Submission",
    "Coding (2)",
    "Bug Fixing",
    "Final Testing",
    "Documentation",
    "Draft Report Submission",
    "Final Project Report Submission",
    "Poster Demonstration & Viva"
]

start_days = [
    0,  # Proposal Submission
    2,  # Requirements Gathering
    5,  # Research on Algorithms
    7,  # UI/UX Design
    12, # Project Flow Design
    14, # Database Design
    17, # System Architecture Design
    20, # Initial Design Submission
    22, # Initial Testing
    24, # Coding (1)
    30, # Bug Fixing (If needed)
    32, # Final Testing
    35, # Professionalism Submission
    37, # Coding (2)
    42, # Bug Fixing
    45, # Final Testing
    48, # Documentation
    51, # Draft Report Submission
    54, # Final Project Report Submission
    56  # Poster Demonstration & Viva
]

durations = [
    2,  # Proposal Submission
    3,  # Requirements Gathering
    2,  # Research on Algorithms
    5,  # UI/UX Design
    2,  # Project Flow Design
    3,  # Database Design
    3,  # System Architecture Design
    2,  # Initial Design Submission
    2,  # Initial Testing
    6,  # Coding (1)
    2,  # Bug Fixing (If needed)
    3,  # Final Testing
    2,  # Professionalism Submission
    5,  # Coding (2)
    3,  # Bug Fixing
    3,  # Final Testing
    3,  # Documentation
    3,  # Draft Report Submission
    2,  # Final Project Report Submission
    4   # Presentation
]

# Create a DataFrame
df = pd.DataFrame({
    'Task': tasks,
    'Start': start_days,
    'Duration': durations,
    'End': np.array(start_days) + np.array(durations)
})


# Reverse the order of tasks in the DataFrame
df_reversed = df.iloc[::-1].reset_index(drop=True)

# Plot the reversed Gantt chart
fig, ax = plt.subplots(figsize=(10, 8))

for i in range(len(df_reversed)):
    ax.broken_barh([(df_reversed.Start[i], df_reversed.Duration[i])], (i-0.4, 0.8), facecolors=('tab:green'))

ax.set_yticks(range(len(df_reversed)))
ax.set_yticklabels(df_reversed.Task)
ax.set_xlabel('Days')
ax.set_title('Healthcare Chatbot Development (8 Weeks)')

plt.show()
```
